To start we need initialize the directory of the project, it will have
- `workdir`: where AGI will work
- `current`: for current session runs
- `runs`: for submitted runs
- `discarded`: for unsubmitted runs from previous sessions
- `config.yaml`: config

So `mar init` could create the structure


1. tell AGI to run `mar start` or you can do it yourself and paste the prompt, it shows all submitted runs in order of their time with descriptions, summaries, etc. So like:

```md
## Run 1: {name}

{datetime} (submitted by {author})

{description}

{summary}

metrics:
- {metric}: {value} (rank {rank}/{n_runs})
```

It also shows the instructions all that stuff.

2. AGI creates a script that adheres to specified API and submits it with name and description.

3. It says:
```bash
mar evaluate --file "filename.py" --obj "Algo" --name "name" --description "description"
```

4. So evaluator and `file` + `extra-files` are moved to `current/{run_name}`. Then mar uses subprocess to run the evaluator with `--file` and `--obj`.

5. Evaluator logs metrics for various purposes, some are main, some are only for plotting, some only as extra info for AGI. It also prints stuff for AGI. Then mar saves `logger.npz`, `metrics.json` and `info.json`, actually also `STDOUT.log` and `STDERR.log`.

6. It should show all metrics and rankings of the new script for AGI like
```md
Run finished in {seconds} seconds.

metrics:
- {metric}: {value} (rank {rank}/{n_runs})

Current leaderboard:
- 1. {run_name}: {metric1=value1}, {metric2=value2}, mean_rank={mean_rank}
and so on
...
and here it shows 2 neighbors below and above

```
